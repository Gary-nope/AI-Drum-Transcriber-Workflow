# OpenDrumScribe: AI Drum Extractor 🥁
This notebook uses the **Demucs v4** model to cleanly extract drum tracks from any song using Google's free cloud GPU. 

**CRITICAL STEP:** Go to `Runtime` > `Change runtime type` > Select `T4 GPU` before running.
**关键步骤：** 运行前请点击顶部菜单栏 `代码执行程序` > `更改附加到运行时的类型` > 选择 `T4 GPU`。

```python
# 1. Install Demucs (Run this cell first / 运行此单元格安装环境)
!python3 -m pip install -U demucs
```

```python
# 2. Upload your song (Run this cell to upload / 运行并上传你的音乐文件)
from google.colab import files
import os
import shutil

# Clear previous uploads to prevent conflicts
if os.path.exists('input_audio'):
    os.remove('input_audio')
    
uploaded = files.upload()
original_file_name = next(iter(uploaded))
file_ext = os.path.splitext(original_file_name)[1]

# Rename file to avoid command line errors with spaces/special characters
safe_file_name = f"input_audio{file_ext}"
os.rename(original_file_name, safe_file_name)
print(f"\nSuccessfully uploaded and prepared: {original_file_name}")
```

```python
# 3. Run AI Separation (This takes 1-2 minutes / 运行 AI 分离，这需要一两分钟)
!demucs "{safe_file_name}"
```

```python
# 4. Download the Drum Track (下载分离出的纯鼓声)
name_no_ext = "input_audio"
drum_path = f"separated/htdemucs/{name_no_ext}/drums.wav"

if os.path.exists(drum_path):
    print("Drum track successfully separated! Downloading...")
    files.download(drum_path)
else:
    print("Error: Drum track not found. Please check the processing logs above.")
```
